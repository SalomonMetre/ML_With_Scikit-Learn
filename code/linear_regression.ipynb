```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([[6],[8],[10],[14],[18]])
y=np.array([7,9,13,17.5,18])
```

```python
plt.title('Pizza prices plotted against diameter')
plt.xlabel('Diameter')
plt.ylabel('Price')
plt.plot(X,y,'.')
plt.axis([0,30,0,30])
plt.grid(True)
plt.show()
```

```python
# model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
test_pizza=np.array([[12]])
predicted_prices=model.predict(test_pizza)
print(f"A 12\" pizza should cost : {predicted_prices[0]}")
```

```python
# plotting the line estimated using linear regression
predicted_prices=model.predict(X)
plt.title("LR line")
plt.xlabel("Pizza diameter")
plt.ylabel("Pizza price")
plt.plot(X,y,'*')
plt.grid(True)
plt.axis([0,30,0,30])
plt.plot(X,predicted_prices,color="green")
plt.show()
predicted_prices
```

```python
# RSS of the model
print(f"The RSS of the model is : {((predicted_prices-y)**2).mean()}")
```

```python
# mean and variance of X
X_mean=X.mean()
X_var=X.var(ddof=1)
print(X_mean, X_var)
```

```python
cov_xy= np.cov(X.transpose(),y)[0][1]
cov_xy
```

```python
# finding the coefficients
beta=cov_xy/X_var
alpha=y.mean()-beta*X_mean
```

```python
def predict_pizza_price(x):
    return alpha+beta*x
```

```python
print(f"The price of a 12\" should cost : {predict_pizza_price(12)}")
```

```python
# testing the model
X_test=np.array([8,9,11,16,12]).reshape(-1,1)
y_test=np.array([11,8.5,15,18,11])
r_squared=model.score(X_test, y_test)
print(f"r-squared : {r_squared}")
```
